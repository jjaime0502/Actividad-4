```python
#Obtenemos un análisis univariado de las vvariables categóricas
freq_tbl(Micro_Retailer)
```

```python
Micro_Retailer.head()
```

```python
#Filtro por columna
dfcategorias=Micro_Retailer.iloc[ :,[9,15,23,24,26,27,34,35,36,48,53,66,67,68,79,82,83,86,88,95]]
```

```python
#Corroboramos valores nulos
valores_nulos=dfcategorias.isnull().sum()
valores_nulos
```

```python
dfcategorias.info()
```

```python
#Eliminamos datos nulos
data_clean_dfcategorias=dfcategorias.copy()
data_clean_dfcategorias=dfcategorias.fillna("SIN RESPUESTA")
data_clean_dfcategorias
```

```python
#Comprobamos datos nulos
valores_nulos=data_clean_dfcategorias.isnull().sum()
valores_nulos
```

##Tabla de variable #1


```python
#Obtengo un análisis univadirado de la variable categórica en específico
table1=freq_tbl(data_clean_dfcategorias['232_type_of_store'])
table1
```

```python
#Obtengo un filtro de los valores más relevantes de la variable categórica seleccionada
Filtro1=table1[table1['frequency']>5]
Filtro1
```

```python
#Ajusto el indice de mi dataframe
Filtro1_index=Filtro1.set_index('232_type_of_store')
Filtro1_index
```

```python
#Realizamos gráfico de barras del dataframe filtrado
Filtro1_index.plot(kind='bar',width=1,figsize=(10,2))
plt.title('Tiendas consultadas')
plt.xlabel('Tipo de tienda')
plt.ylabel('Frecuencia')
```

Con esta tabla se logra visualizar que los tipos de tienda que tuvieron mayor presencia en la escuesta es de Microrestaurantes y Tienda de comestibles
